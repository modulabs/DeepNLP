```python
import sklearn
sklearn.__version__
```

```python
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
print("iris_dataset key: \n{}".format(iris_dataset.keys()))
```

```python
print(iris_dataset.DESCR)
print(iris_dataset.target_names)
```

```python
# count row & column 
iris_dataset.data.shape
```

```python
print(iris_dataset.target)
iris_dataset.target_names
```

```python
iris_dataset.feature_names
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_label, test_label = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=42)
```

```python
# 훈련 데이터의 문제와 답 건수 확인
print(train_input.shape)
train_label.shape
```

```python
# 테스트 데이터의 문제와 답 건수 확인
print(test_input.shape)
test_label.shape
```

```python
print(test_input[:5])
test_label[:5]
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(train_input, train_label)
```

```python
import numpy as np
new_input = np.array([[6.1, 2.8, 4.7, 1.2]])
new_input
```

```python
prediction = knn.predict(new_input)
prediction
```

```python
# 예측된 모델 평가하기
eval_label = knn.predict(test_input)
eval_label
```

```python
print('test accurance {:.2f}'.format(np.mean(eval_label == test_label)))
```
